```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#(a) Line plot
x = np.arange(1,6*np.pi,0.6) # start,stop,step
y = np.sin(x) # sin() function
z = np.cos(x) #cos() function
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi') # X Label
plt.ylabel('sin(x) and cos(x)') # Y Label
plt.title('Plot of sin and cos from 0 to 4pi') #adding title
plt.legend(['sin(x)', 'â€˜cos(x)']) # Legends
plt.show()
```

```python
#(b) bar plot
# creating the dataset
sub_data = {'Machine Learning':7, 'Artificial Intelligence':16, 'DataMining':10,"Information security":12,"Computer graphics":30}
stu_subjects = list(sub_data.keys())
students = list(sub_data.values())
fig = plt.figure(figsize = (20, 10))
# creating the bar plot
plt.bar(stu_subjects, students, color ='red', width = 0.5)
plt.xlabel("Subjects Offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different Subjects in 6 semester")
plt.show()
```

```python
#(c) scatter plot
x = np.random.normal(2.0, 8.0, 2000)
y = np.random.normal(16.0, 31.0, 2000)
plt.scatter(x, y)
plt.show()
```

```python
#(d) pte plot
plt.pie(students, labels = stu_subjects)
plt.legend(title = "Five _subjects",loc="lower right")
plt.show()
```

```python
#(e)area plot
# Create data
x = range(1,10)
y = [4,2,5,3,6,8,3,9,7]
# Area plot
plt.fill_between(x, y,color ='blue')
plt.show()
```

```python
#Generate different subplots from a given plot and color plot data
# Data for plotting
x = np.arange(0.0, 2.0, 0.08)
y = 1 + np.cos(2 * np.pi * x)
# Creating 6 subplots and unpacking the output array
fig, ((ax1,ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2)
ax1.plot(x, y, color="lightgreen")
ax2.plot(x, y, color="orange")
ax3.plot(x, y, color="black")
ax4.plot(x, y, color="yellow")
ax5.plot(x, y, color="magenta")
```
